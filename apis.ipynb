### Introducción
Dentro de las técnicas de obtención de datos hemos estudiado los repositorios, el web scraping y la extracción de información usando expresiones regulares.

Ahora veremos un servicio que nos ofrecen muchas plataformas para facilitarnos la vida a la hora de llevar a cabo esa obtención de información, las APIs.

### Introducción

API significa "application programming interface" (interfaz de programación de aplicaciones). En pocas palabras, una API consiste en un conjunto de reglas que describen cómo una aplicación puede interactuar con otra, y los mecanismos que permiten que dicha interacción ocurra.

¿Qué es una interacción entre dos aplicaciones? Típicamente, una interacción ocurre cuando una aplicación quiere acceder a los datos que tiene otra aplicación, o enviar datos a esa aplicación. Otra interacción podría ser cuando una aplicación quiere solicitar un servicio a otra.

### Introducción
Una API no es (normalmente) una interfaz de usuario. Proporciona interacción de software a software, no de usuario. A veces, sin embargo, una API puede proporcionar un widget de interfaz de usuario, que una aplicación puede agarrar y mostrar.

Hay dos beneficios principales que aporta una API:

1. Simplificación, al proporcionar una capa que oculta la complejidad.
2. Estandarización.

### Introducción
Ejemplos:
> Microsoft Word pide a la impresora activa que devuelva su estado. A Microsoft Word no le importa qué tipo de impresora está disponible. La API se preocupa por eso.

> Los bloggers en WordPress pueden incrustar su flujo de Twitter en la barra lateral de su blog. WordPress usa la API de Twitter para habilitar esto.

### APIs de servicios web
Un servicio web es una pieza de software, o un sistema, que proporciona acceso a sus servicios a través de una dirección en la World Wide Web. Esta dirección se conoce como URI, o URL. El punto clave es que el servicio web ofrece su información en un formato que otras aplicaciones pueden "entender" o analizar.

Ejemplos:
> La API de Flickr, la API de Google Static Maps y los demás servicios web de Google Maps.

### APIs de servicios web
Un servicio web utiliza HTTP para intercambiar información. (O HTTPS, que es una versión encriptada de HTTP.)

Cuando una aplicación, el "cliente", quiere comunicarse con el servicio web, la aplicación envía una petición HTTP. El servicio web envía entonces una respuesta HTTP.

En la solicitud, gran parte de la información requerida se pasa en la propia URL, como rutas en la URL y/o como parámetros de la URL.

Por ejemplo:
> http://maps.googleapis.com/maps/api/staticmap?center=Sydney,NSW&zoom=14&size=400x400&sensor=false

### APIs de servicios web
Además de la URL, las solicitudes y respuestas HTTP incluirán información en el encabezado y el cuerpo del mensaje. Las "cabeceras" de solicitudes y respuestas incluyen varios tipos de metadatos, como el navegador que se utiliza, el tipo de contenido, el idioma (humano, no software), y más.

El organismo incluye datos adicionales en la solicitud o respuesta. Los formatos de datos más comunes son XML y JSON. El proceso de conversión de datos del formato interno (por ejemplo, una base de datos o una clase) al formato transferible se denomina "serialización de datos".

Tipos de servicios web más utilizados:

- SOAP
- XML-RPC
- JSON-RPC
- REST

### SOAP (Simple Object Access Protocol)
SOAP es un protocolo que define el método de comunicación y la estructura de los mensajes. El formato de transferencia de datos es XML.

Un servicio SOAP publica una definición de su interfaz en un documento legible por máquina, utilizando [WSDL - Web Services Definition Language](https://www.w3.org/2001/04/wsws-proceedings/uche/wsdl.html).

### SOAP (Simple Object Access Protocol)

Petición:

```xml
<?xml version="1.0"?>

<soap:Envelope
xmlns:soap="http://www.w3.org/2003/05/soap-envelope/"
soap:encodingStyle="http://www.w3.org/2003/05/soap-encoding">

<soap:Body>
  <m:GetPrice xmlns:m="https://www.w3schools.com/prices">
    <m:Item>Apples</m:Item>
  </m:GetPrice>
</soap:Body>

</soap:Envelope>
```

### SOAP (Simple Object Access Protocol)

Respuesta:

```xml
<?xml version="1.0"?>

<soap:Envelope
xmlns:soap="http://www.w3.org/2003/05/soap-envelope/"
soap:encodingStyle="http://www.w3.org/2003/05/soap-encoding">

<soap:Body>
  <m:GetPriceResponse xmlns:m="https://www.w3schools.com/prices">
    <m:Price>1.90</m:Price>
  </m:GetPriceResponse>
</soap:Body>

</soap:Envelope>
```

### XML-RPC
XML-RPC es un protocolo más antiguo que SOAP. Utiliza un formato XML específico para la transferencia de datos, mientras que SOAP permite un formato XML propietario.

Una llamada XML-RPC tiende a ser mucho más simple y a utilizar menos ancho de banda que una llamada SOAP. (SOAP es conocido por ser "verboso".) SOAP y XML-RPC tienen diferentes niveles de soporte en varias bibliotecas.

### XML-RPC

Petición:

```xml
<?xml version="1.0" encoding="ISO-8859-1"?>
<methodCall>
   <methodName>sample.sum</methodName>
   <params>
      <param>
         <value><int>17</int></value>
      </param> 
      <param>
         <value><int>13</int></value>
      </param>
   </params>
</methodCall>
```

### XML-RPC

Respuesta:

```xml
<?xml version="1.0" encoding="ISO-8859-1"?>
<methodResponse>
   <params>
      <param>
         <value><int>30</int></value>
      </param>
   </params>
</methodResponse>
```

### JSON-RPC
JSON-RPC es similar a XML-RPC, pero utiliza JSON en lugar de XML para la transferencia de datos.

Petición:
```json
{"jsonrpc": "2.0", "method": "subtract", "params": {"minuend": 42, "subtrahend": 23}, "id": 3}
```

Respuesta:
```json
{"jsonrpc": "2.0", "result": 19, "id": 3}
```

### REST (Transferencia del estado de representación)

REST no es un protocolo, sino un conjunto de principios arquitectónicos. Lo que diferencia a un servicio REST de otros servicios web es su arquitectura.

Algunas de las características requeridas de un servicio REST incluyen la simplicidad de las interfaces, la identificación de los recursos dentro de la solicitud y la capacidad de manipular los recursos a través de la interfaz.

### REST (Transferencia del estado de representación)

Desde el punto de vista de una aplicación cliente, los servicios REST tienden a ofrecer una estructura URL fácil de usar, compuesta principalmente de sustantivos que reflejan las categorías lógicas y jerárquicas de los datos que se ofrecen.

### REST (Transferencia del estado de representación)
Por ejemplo, supongamos que necesita obtener una lista de árboles de una API en example-tree-service.com. Se puede presentar una solicitud como esta:
> http://example-tree-service.com/trees

Tal vez usted ya conoce el nombre científico de una familia de árboles, `Leptospermum`, y necesita conocer el nombre común. Su petición podría parecerse a ésta:
> http://example-tree-service.com/trees/leptospermum


El servicio de árboles podría entonces enviar una respuesta que contenga un montón de información sobre la familia `Leptospermum`, incluyendo un campo "common-name" que contenga el valor "teatrees".

### REST (Transferencia del estado de representación)

El formato de datos más utilizado es JSON o XML. A menudo el servicio ofrece una opción, y el cliente puede solicitar una u otra incluyendo "json" o "xml" en la ruta de la URL o en un parámetro de la URL.

Un servicio REST puede publicar un documento WADL que describa los recursos que tiene disponibles y los métodos que aceptará para acceder a ellos. WADL significa Web Application Description Language (Lenguaje de descripción de aplicaciones web). Es un formato XML que proporciona una descripción procesable por máquina de las aplicaciones Web basadas en HTTP.

Si no hay ningún documento de WADL disponible, los desarrolladores confían en la documentación para informarles sobre los recursos y métodos disponibles. La mayoría de los servicios web siguen dependiendo de la documentación más que de una descripción legible por máquina de su interfaz.

### REST (Transferencia del estado de representación)

En un servicio REST bien definido, no existe un acoplamiento entre la interfaz REST y la arquitectura subyacente del servicio. Esto se cita a menudo como la principal ventaja de REST sobre las arquitecturas RPC (Remote Procedure Call).

Los clientes que llaman al servicio no dependen de los nombres de métodos o estructuras de datos subyacentes del servicio. En su lugar, las interfaces REST sólo representan los recursos lógicos y la funcionalidad disponibles.

La estructura de los datos del mensaje es independiente de la estructura de datos del servicio. El mensaje contiene una representación de los datos. Los cambios en el servicio subyacente no deben romper a los clientes.

### GraphQL

- En ocasiones solo se necesita un único campo de un recurso, pero el servidor devuelve todos, sin embargo, con GraphQL es la aplicación la que tiene el control del recurso, reduciendo el ancho de banda y los tiempos.
- Con GraphQL se puede acceder a distintos recursos en una única llamada.
- Se elimina el concepto de endpoints, sustituyéndolo por tipos con los indicar solo lo que se necesita.
- Se intenta eliminar el versionado de las APIs pudiendo deprecar el uso de atributos. 
- GraphQL desacopla la dependencia con el acceso a las bases de datos.

### GraphQL
```json
schema {
    query: Query
    mutation: Mutation
}
 
type Query {
    pizzas (name: [String]): [Pizza]
    ingredients: [Ingredient]
}
 
type Mutation {
    createPizzas (pizzas: [PizzaInput]): [Pizza]
}
```

### GraphQL

```json
type Pizza {
    id: Int!
    name: String!
    origin: String
    ingredients: [Ingredient]
}

type Ingredient {
    id: Int!
    name: String!
    calories: String
}

input PizzaInput {
    name: String!
    origin: String
    ingredientIds: [Int]
}
```

### Tipos de API

Acabamos de ver en las diapositivas anteriores dos formas de acceder a los datos de una API. Ahora necesitamos saber qué necesitamos para acceder a esos datos.

Centrándonos en la necesidad de proporcionar credenciales para acceder a una API, podemos dividir los tipos de API en dos: con y sin autenticación. Para trabajar con APIs que necesiten autenticación necesitaremos unos credenciales para poder acceder e interactuar con el servicio. En el otro tipo no necesitamos ningún tipo de información extra.

### Tipos de API - Sin autenticación
Hoy en dı́a, la mayorı́a de las APIs necesitan que nos autentiquemos para poder usarlas. Sin embargo hay casos (o versiones de la API) en los que podemos acceder a ciertos endpoints de una API sin necesidad de autenticación alguna.

Lógicamente, este tipo de acceso reduce las prestaciones del servicio, ya sea en número de peticiones o en la calidad del mismo.

En el siguiente punto del tema veremos un par de ejemplos de acceso a APIs de este tipo.

### Tipos de API - Sin autenticación

````python
import requests
import json
endpoint = "https://api.github.com/users/versae/repos"
repos = json.loads(requests.get(endpoint).text)
repos[5]
````

### Tipos de API - Con autenticación

En el caso de las APIs con autenticación necesitaremos de unos credenciales para su acceso. Generalmente, estos credenciales los obtendremos registrándonos en la plataforma que ofrece el servicio.

Aquı́ también podemos encontrarnos que el registro puede ser gratuito o de pago. Nuevamente tendremos que analizar los recursos que necesitamos para saber qué tipo de cuenta necesitaremos crear.

Como ejempo de API de este tipo, veremos un ejemplo de la API de Twitter en la siguiente sección.

### Tipos de API - Con autenticación

auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)

### Construyendo APIs - OpenAPI

[Connexion](https://connexion.readthedocs.io/en/latest/) es un framework Python para transformar APIs especificadas en OpenAPI a esqueletos de servidores REST.

```python

import connexion

options = {"swagger_ui": False}
app = connexion.FlaskApp(__name__, specification_dir='openapi/',
                    options=options)
app.add_api('my_api.yaml')
```

> Ejemplo: http://62.204.199.36:5000/openapi.json (UI: http://62.204.199.36:5000/ui)

### Construyeno APIs - Hug

[Hug](https://www.hug.rest/) es un framework para construir módulos Python, aplicaciones de línea de comandos, e interfaces REST maximizando la reutilización de código.


- Acceso local

```python
"""First API, local access only"""
import hug


@hug.local()
def happy_birthday(name: hug.types.text, age: hug.types.number, hug_timer=3):
    """Says happy birthday to a user"""
    return {'message': 'Happy {0} Birthday {1}!'.format(age, name),
            'took': float(hug_timer)}
```

### Construyeno APIs - Hug

- Acceso HTTP

"""First hug API (local and HTTP access)"""

### Construyeno APIs - Hug

- Línea de comandos

```python
"""First hug API (local, command-line, and HTTP access)"""
import hug


@hug.cli()
@hug.get(examples='name=Timothy&age=26')
@hug.local()
def happy_birthday(name: hug.types.text, age: hug.types.number, hug_timer=3):
    """Says happy birthday to a user"""
    return {'message': 'Happy {0} Birthday {1}!'.format(age, name),
            'took': float(hug_timer)}


if __name__ == '__main__':
    happy_birthday.interface.cli()
```

### Construyendo APIs - FastAPI

```python
from fastapi import FastAPI

app = FastAPI()


@app.get("/")
def read_root():
    return {"Hello": "World"}


@app.get("/items/{item_id}")
def read_item(item_id: int, q: str = None):
    return {"item_id": item_id, "q": q}
```

### Conclusiones
Las APIs nos ofrecen una serie de recursos (endpoints) que nos facilitan las tareas de obtención de datos. Es uno de los servicios más usados a la hora de obtener datos de terceros.

Hemos visto cómo acceder a APIs con y sin autenticacion y cómo usarlas por medio de urls y de librerı́as de lenguajes de programación.

Es necesario conocer y trabajar con la documentación que ofrece una API para saber qué endpoints podemos usar, cómo usarlos, condiciones de uso, etc.
