### Introducción
Dentro de las técnicas de obtención de datos hemos estudiado los repositorios, el web scraping y la extracción de información usando expresiones regulares.

Ahora veremos un servicio que nos ofrecen muchas plataformas para facilitarnos la vida a la hora de llevar a cabo esa obtención de información, las APIs.

### Introducción

Si queremos dar una definición formal, podrı́amos decir que una API (del inglés Application Programming Interface) es un conjunto de funciones y procedimientos que ofrece una biblioteca para ser utilizado por otro software como una capa de abstracción.

Desde un punto de vista sencillo y breve, podrı́amos ver una API como una serie de funciones que una plataforma pone a nuestra disposición para poder usarlas en nuestras aplicaciones.

### Introducción
Para nosotros, esas funciones serán servicios que ciertas plataformas nos ofrecen para obtener datos. Esto, por ejemplo, nos ahorrará tiempo y esfuerzo a la hora de evitar trabajos más tediosos como realizar scraping.

Generalmente, las APIs se comunican con los usuarios usando texto plano. Debido a esto, los datos que obtendremos por medio de las APIs usarán algún tipo de formato de datos de los que hemos estudiados, siendo JSON el más extendido.

### Introducción
Es común acompañar una API con documentación relacionada con los servicios que ofrece. Esta documentación nos indicará cómo y de qué forma podemos acceder a los recursos de la API, el ratio de llamadas permitidas, si necesitamos autenticación o no, etc. Los recursos que nos ofrece una API son los que se conocen como
endpoints.

Para ver qué es un endpoint de una forma sencilla: son urls a las que nos conectaremos y nos devolverán la información que deseamos. Hay que tener en cuenta que dependiendo del endpoint necesitaremos credenciales de conexión, usar un método u otro para hacer la petición, etc.

### Introducción
Aunque el uso de APIs nos proporciona muchos factores positivos, también tenemos que tener en cuenta algunos negativos. Al usar APIs (y en nuestro caso, servicios que ofrecen las compañı́as) nos
exponemos a ciertos problemas como:
- Condiciones cambiantes del uso del servicio
- Disponibilidad del servicio
- Falta de innovación del servicio utilizado
- Privacidad (muy dependiente del paı́s)

### Formas de acceso a una API
Hemos visto en el punto anterior que las APIs con las que trabajaremos nos ofrecen sus servicios por medio de endpoints.

También hemos visto que un endpoint no deja de ser una url que nos devuelve cierto tipo de información.

Tenemos muchas formas de acceder a la información que nos proporciona un endpoint. Podemos hacerlo directamente por el navegador, usando un programa que ejecute peticiones, usando un paquete que nos conecte directamente con la API, etc. Veamos unos ejemplos en las siguientes diapositivas.

### Formas de acceso a una API

Como hemos comentado, en el caso de que accedamos a nuestra API por medio de urls podemos usar un simple navegador web para probar nuestra API o usar un paquete que nos permita realizar
peticiones:

````python
import requests
import json
endpoint = "https://api.github.com/users/versae/repos"
repos = json.loads(requests.get(endpoint).text)
repos
````

### Tipos de API

Acabamos de ver en las diapositivas anteriores dos formas de acceder a los datos de una API. Ahora necesitamos saber qué necesitamos para acceder a esos datos.

Centrándonos en la necesidad de proporcionar credenciales para acceder a una API, podemos dividir los tipos de API en dos: con y sin autenticación. Para trabajar con APIs que necesiten autenticación necesitaremos unos credenciales para poder acceder e interactuar con el servicio. En el otro tipo no necesitamos ningún tipo de información extra.

### Tipos de API - Sin autenticación
Hoy en dı́a, la mayorı́a de las APIs necesitan que nos autentiquemos para poder usarlas. Sin embargo hay casos (o versiones de la API) en los que podemos acceder a ciertos endpoints de una API sin necesidad de autenticación alguna.

Lógicamente, este tipo de acceso reduce las prestaciones del servicio, ya sea en número de peticiones o en la calidad del mismo.

En el siguiente punto del tema veremos un par de ejemplos de acceso a APIs de este tipo.

### Tipos de API - Sin autenticación

````python
import requests
import json
endpoint = "https://api.github.com/users/versae/repos"
repos = json.loads(requests.get(endpoint).text)
repos
````

### Tipos de API - Con autenticación

En el caso de las APIs con autenticación necesitaremos de unos credenciales para su acceso. Generalmente, estos credenciales los obtendremos registrándonos en la plataforma que ofrece el servicio.

Aquı́ también podemos encontrarnos que el registro puede ser gratuito o de pago. Nuevamente tendremos que analizar los recursos que necesitamos para saber qué tipo de cuenta necesitaremos crear.

Como ejempo de API de este tipo, veremos un ejemplo de la API de Twitter en la siguiente sección.

### Tipos de API - Con autenticación
```python
# Twitter
auth = twitter.oauth.OAuth(
    OAUTH_TOKEN,
    OAUTH_TOKEN_SECRET,
    CONSUMER_KEY,
    CONSUMER_SECRET
)
twitter_api = twitter.Twitter(auth=auth)
# Twitter API ready to use...
```

Para acceder a la información de esta platafaforma usaremos una API gratuita llamada OMDb API. Aunque la API sea gratuita, necesitaremos obtener un API key para poder realizar nuestras
peticiones.

OMDb API nos ofrece dos tipos de endpoints para obtener información dependiendo de si buscamos por tı́tulo, id o haciendo una búsqueda. La información que nos devuelve es la misma que podemos encontrar en IMDB. Los parámetros que podemos usar en los endpoints y la documentación de esta API la podemos encontrar en: http://www.omdbapi.com/

### Ejemplos - IMDB

````python
# IMDB
import requests
import json
# By title
url ="http://www.omdbapi.com/?t=matrix&apikey=*******"
content = requests.get(url).json()
print(content)
````

````python
# By search
url ="http://www.omdbapi.com/?s=star wars&apikey=*******"
content = requests.get(url).json()
print(content)
````

````python
# By id
url ="http://www.omdbapi.com/?i=tt0133093&apikey=*******"
content = requests.get(url).json()
print(content)
````

### Ejemplos - Twitter
Una de las APIs que más se usan actualmente es la de Twitter. Una de las razones es que el tipo de información que devuelve es muy caracterı́stica, ya que cada unidad de datos (tweet) es de un
tamaño reducido.

Principalmente, Twitter ofrece dos tipos de servicios, una API clásica donde pedimos los últimos tweets relacionados con nuestra búsqueda (REST API) o una API en tiempo real (Streaming API) que nos permite ir recolectando los tweets que se van produciendo sobre una búsqueda determinada. Como nota importante, la API de Twitter ha cambiado mucho en sus últimas versiones, para más información: https://developer.twitter.com/en/docs
